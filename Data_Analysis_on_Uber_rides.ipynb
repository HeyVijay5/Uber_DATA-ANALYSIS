Data Preprocessing


```python
dataset['PURPOSE'].fillna("NOT", inplace = True)
```

```python
dataset.head()
```

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce')
dataset['END_DATE'] = pd.to_datetime(dataset['END_DATE'], errors = 'coerce')
```

```python
dataset.info()
```

```python
from datetime import datetime
dataset['date'] = pd.DatetimeIndex(dataset['START_DATE']).date
dataset['time'] = pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head()
```

```python
dataset['day-night'] = pd.cut(x=dataset['time'],bins = [0,10,15,19,24],labels = ['Night', 'Morning', 'Afternoon', 'Evening'],  )
```

```python
dataset.head()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.shape
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'dataset' is already defined as a DataFrame

# Plotting
plt.figure(figsize=(20, 5))

# First subplot for CATEGORY
plt.subplot(1, 2, 1)
sns.countplot(x=dataset['CATEGORY'])  # Use x=dataset['CATEGORY'] for countplot
plt.xticks(rotation=90)
plt.title('Count of Categories')

# Second subplot for PURPOSE
plt.subplot(1, 2, 2)
sns.countplot(x=dataset['PURPOSE'])  # Use x=dataset['PURPOSE'] for countplot
plt.xticks(rotation=90)
plt.title('Count of Purposes')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
dataset['MONTH'] = pd.DatetimeIndex(dataset['START_DATE']).month# START_DATE se
month_label = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'April',
 5.0: 'May', 6.0: 'June', 7.0: 'July', 8.0: 'Aug',
 9.0: 'Sep', 10.0: 'Oct', 11.0: 'Nov', 12.0: 'Dec'}
```

```python
dataset["MONTH"] = dataset.MONTH.map(month_label) # Number months ko string name
mon = dataset.MONTH.value_counts(sort=False)
```

```python
dataset.head()
```

```python
df = pd.DataFrame({
 "MONTHS": mon.values, # Har month ka total count.
 "VALUE COUNT": dataset.groupby('MONTH', sort=False)['MILES'].max() # Har mo
})
p = sns.lineplot(data=df) # Line plot banata hai.
p.set(xlabel="MONTHS", ylabel="VALUE COUNT")
```
